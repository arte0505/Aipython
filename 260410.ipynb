```python
fruits=['banana', 'orange', 'grape', 'mango']
```

```python
fruits.extend(['kiwi', 'peach', 'pineapple'])
```

```python
len(fruits)
```

```python
for f in fruits:
  print(f)
```

```python
fruits_upper = []
for f in fruits:
    fruits_upper.append(f.upper())
    print(fruits_upper)
```

```python
fruits_len6 = []               # 빈 리스트 준비
for f in fruits:               # fruits에서 과일을 하나씩 꺼내서
    if len(f) >= 6:            # 글자 수가 6 이상이면
        fruits_len6.append(f)  # 새 리스트에 추가
        print(fruits_len6)
```

```python
for i, fruit in enumerate(fruits):
  i+=1
  print(f"{i},{fruit}")
```

```python
king = [
    "태조", "정종", "태종", "세종", "문종", "단종", "세조", "예종", "성종",
    "연산군", "중종", "인종", "명종", "선조", "광해군", "인조", "효종", "현종", "숙종",
    "경종", "영조", "정조", "순조", "헌종", "철종", "고종", "순종"
]
```

```python
len(king)
```

```python
print("세종" in king)
print("태상왕" in king)
```

```python
pos=king.index("세종")
print(f"{pos+1}번째")
```

```python
name = input()
if name in king:
  pos = king.index(name)
  print(f"{pos+1}번째 왕입니다")
else:
  print('존재하지 않습니다')
```

```python
reversed_king = list(reversed(king))
print(reversed_king)
```
