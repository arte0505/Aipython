```python
# 사용자에게 숫자를 입력받아 짝수/홀수를 출력하는 프로그램

num = int(input("숫자를 입력하세요: "))

# % 는 나머지를 구하는 연산자입니다.
# 어떤 수를 2로 나눴을 때 나머지가 0이면 짝수, 1이면 홀수입니다.
if num > 0:
    print("양수입니다")
else:
    print("음수입니다")
```

```python
cf=2000
la=3000
g=2500
j=2800

print('주문금액:',
(int(input('커피 주문수'))*cf)+
(int(input('라떼 주문수'))*la)+
(int(input('녹차 주문수'))*g)+
(int(input('주스 주문수'))*j),'원입니다')

```

```python
cf = 2000
la = 3000
g  = 2500
j  = 2800

coffee_cnt = int(input("커피 주문수: "))
latte_cnt  = int(input("라떼 주문수: "))
green_cnt  = int(input("녹차 주문수: "))
juice_cnt  = int(input("주스 주문수: "))

total = coffee_cnt * cf + latte_cnt * la + green_cnt * g + juice_cnt * j
print(f"주문금액: {total}원입니다")
```
