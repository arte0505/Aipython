```python
# 사칙연산(+, -, *, /)을 변수와 연산자, 출력문만으로 보여주는 아주 간단한 예제

a = 10        # 첫 번째 숫자를 변수 a에 저장해두면, 여러 연산에 같은 값을 재사용할 수 있어요
b = 3         # 두 번째 숫자를 변수 b에 저장해두면, a와 마찬가지로 반복해서 사용할 수 있어요

print(a + b)  # 덧셈(+) 결과를 화면에 출력해서 “a+b가 얼마인지” 확인해요
print(a - b)  # 뺄셈(-) 결과를 화면에 출력해서 “a-b가 얼마인지” 확인해요
print(a * b)  # 곱셈(*) 결과를 화면에 출력해서 “a*b가 얼마인지” 확인해요
print(a / b)  # 나눗셈(/) 결과를 화면에 출력해서 “a/b가 얼마인지” 확인해요 (파이썬은 보통 실수로 나와요)
```

```python
# 사용자에게서 두 수를 입력받아 사칙연산 결과를 출력하는 프로그램

a = int(input("첫 번째 수를 입력하세요: "))  # 키보드로 입력받은 값은 글자(문자열)라서, 계산하려면 정수로 바꿔요
b = int(input("두 번째 수를 입력하세요: "))  # 두 번째 값도 같은 이유로 정수로 바꿔요

print(a + b)  # 덧셈 결과 출력
print(a - b)  # 뺄셈 결과 출력
print(a * b)  # 곱셈 결과 출력
print(a / b)  # 나눗셈 결과 출력 (b가 0이면 오류가 날 수 있어요)
```
