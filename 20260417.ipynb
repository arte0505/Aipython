```python
# 메뉴판 데이터
cafe_menu = {
    "아메리카노": 4000, "카페라떼": 4500, "카푸치노": 4500,
    "바닐라라떼": 5000, "카페모카": 5000, "에스프레소": 3500,
    "콜드브루": 4500, "자몽에이드": 5500, "레몬티": 5000, "초코라떼": 4800
}

order_history = {}
total_price = 0

print("=== [카페 메뉴판] ===")
for menu, price in cafe_menu.items():
    print(f"{menu}: {price}원")
print("=====================")

print("\n주문을 시작합니다 (종료하려면 '종료'를 입력하세요)")

while True:
    order = input("\n주문할 메뉴를 입력하세요: ")

    if order == "종료":
        break

    if order in cafe_menu:
        # [핵심 수정] try-except로 숫자 입력 확인
        try:
            count = int(input(f"{order} 몇 잔을 주문하시겠습니까? "))
        except ValueError:
            print("잘못된 입력입니다. 숫자(정수)만 입력해주세요.")
            continue # 다시 주문 단계로 돌아감

        # 수량 검증
        if count <= 0:
            print("잘못된 수량입니다. 1잔 이상 입력해주세요.")
            continue

        # 주문 내역 저장
        if order in order_history:
            order_history[order] += count
        else:
            order_history[order] = count

        total_price += cafe_menu[order] * count
        print(f"{order} {count}잔이 추가되었습니다.")
    else:
        print("죄송합니다. 메뉴판에 없는 메뉴입니다.")

# 최종 주문 내역 출력
if order_history:
    summary_list = [f"{menu} {count}잔" for menu, count in order_history.items()]
    summary_str = ", ".join(summary_list)

    print(f"\n주문이 완료되었습니다.")
    print(f"{summary_str}, 총 결제 금액은 {total_price}원입니다.")
else:
    print("\n주문하신 내역이 없습니다.")
```
